```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_train.ndim
```

```python
model.intercept_
# When experience is 0 the salary will be 26801.96951300559
```

```python
model.coef_
# For every one year the salary will be increased by 9216.41840266
```

# model.predict is used in two scenarios

    1. To evaluate how well the model learns the pattern or Evaluation
    
    2. Future predictions

-- Erros(model mistake) -> Actual output - predicted output

-- predict() -> model.predict(x_train) -> Takes x_train and provide y predict

-- model.score(x,y)

```python
predicted_y_train=model.predict(x_train)

error=y_train - predicted_y_train
error
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.scatter(x_train,predicted_y_train)
```

```python
pred_y_test=model.predict(x_test)

error=y_test - pred_y_test
error
```

```python
plt.scatter(x_test,y_test)
plt.scatter(x_test,pred_y_test)
```

model.score() -> It is used to evaluate performance of the model

```python
model.score(x_train,y_train)*100
```

```python
model.score(x_test,y_test)*100
```
