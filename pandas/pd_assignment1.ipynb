```python
# # Convert Size,installs,price to integer
def convert(value):
    value=value.replace('M','')
    value=value.replace('+','')
    value=value.replace('k','')
    value=value.replace(',','')
    if value=='Varies with device':
        value=None
    return value


# def changes(value):
#     if 'M' in value:
#         return value.removesuffix('M')
#     if value=='Varies with device':
#         return None
#     if "+" in value:
#         return value.removesuffix('+')
#     if "k" in value:
#         return value.removesuffix('k')
#     if ',' in value:
#         return value.replace(',','')
```

# Run All Once Again **

```python
df['Size']=df['Size'].apply(convert)
df['Size']=df['Size'].astype(float)

# df['Size']=df['Size'].apply(changes)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Installs']=df['Installs'].str.replace('+','').str.replace(',','')
df['Installs']=df['Installs'].replace({'Free':None})
df['Installs']=df['Installs'].astype(float)
```

```python
df['Price']=df['Price'].str.replace('$','')
df['Price']=df['Price'].replace({'Everyone':None})
df['Price']=df['Price'].astype(float)
```
