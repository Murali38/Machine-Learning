# AGG()
Syntax 1: Gives min,max and other values to all columns df.agg(['func1','func2',...'funcN'])

Syntax 2: Calculate seperate functions for rach columns
df.agg({
    'col_name':'func1',
    'col_name':'func2'
})

```python
df[['reading score','writing score','math score']].agg(['mean','max','min'])
```

# Second Syntax

```python
df.agg({
    'writing score':'mean',
    'reading score':'max',
    'math score':'min'
})
```
