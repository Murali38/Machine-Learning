Syntax -> df.groupby('col_names')

# MAP() and REPLACE ()

```python
# map() and replace() are used to replace the values

# map is used to only replace value from a specific column

# replace is used to replace values in an entire dataframe

# syntax -> df["colname"].map({'oldvalue':'newvalue'}

# syntax -> df.replace({'oldvalue':'newvalue'})
# df['colname'].replace({'old value':'new value'})
```

```python

df['gender'].map({'female':'F'})
# replaces only the mentioned value and keeps other value as NaN
```

```python
df['gender'].map({'female':'F','male':'M'})
```

```python
df['gender'].replace({'female':'F'})
# replaces female name as F and male as male
```

```python
df.replace({'female':'F'})
```
