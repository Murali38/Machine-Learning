Outliers:
    1.z-score -> x-mean/std

      - High variance means the actual values are far away from mean variance gives the average of x's squared units which are far away from actual values but standard deviation tells the actual units hence std is recommended

    2. IQR
    
    3.boxplot

# Variance = summation((xi-mean(x))^2)/n

xi=Actual Values

z-score -> (x-mue)/std is used for two reasons

    1.Scalin(ML preprocessing Technique)

    2.To find outliers

How to Find Outliers ?

 - First apply z-score method to the required column
 
 - After applying the z-score method if the value is above3 or less than -3 those are outliers

```python
def z_score(col):
    X=df1[col]
    x_mean=df1[col].mean()
    x_std=df1[col].std()
    return(X-x_mean)/x_std
df1['Rating_z']=z_score('Rating')
df1['Rating_z']
```

```python
def outliers(col='Rating_z'):
    l=[]
    for val in df1['Rating_z']:
        if abs(val)>3:
            l.append(val)
    return l
    
outliers()
```

```python
len(outliers())
```

```python
df1=df1.dropna()
```

```python
plt.boxplot(df1['Rating_z'])
```
