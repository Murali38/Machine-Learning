```python
import pandas as pd
import numpy as np
```

```python
# np.random.seed(42) # Set random seed for reproducibility
# n_rows=1000 # Generate 1000 rows with 4 columns
```

```python
np.random.seed(42)

n_rows = 1000
data = {
    'color': np.random.choice(['red', 'blue', 'green'], n_rows),
    'size': np.random.choice(['small', 'medium', 'large'], n_rows),
    'material': np.random.choice(['plastic', 'metal', 'wood'], n_rows),
    'price': np.random.randint(10, 500, n_rows)
}
df = pd.DataFrame(data)

# Display first few rows
print(df.head())
print(f"\nDataset shape: {df.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
from category_encoders import TargetEncoder
```

```python
x=df[['color','size','material']]
y=df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
label_encoder=LabelEncoder()
```

```python
# while doing label_encoder.fit_transform we cannot pass multiple columns
```

```python
updated_xtrain=label_encoder.fit_transform(x_train['color'])
updated_xtrain
```

```python
updated_xtest=label_encoder.fit_transform(x_test['color'])
updated_xtest
```

```python
# LabelEncoder can't handle unknown values
# label_encoder.transform(['brown']) cannot be done as it can't handle unknown values (unseen data)
```

# OneHotEncoder()

```python
onehot_encoder=OneHotEncoder(sparse_output=False)
# By default sparse_output is True
# sparse_output is used to conver into array by giving sparse_output as False

# or we can do 

# values=values.toarray() to convert into array
```

```python
values=onehot_encoder.fit_transform(x_train)
values
```

```python
# onehot_encoder.transform(xtest[['color','size']])
# we need to pass all the feature columns otherwise it raise ValueError 
```

```python
onehot_encoder.transform(x_test)
```

```python
# onehot_encoder.transform([['red','large'],['blue','small']])
# ValueError: X has 2 features, but OneHotEncoder is expecting 3 features as input
```

```python
onehot_encoder.transform([['red','large','plastic'],['blue','small','wood']])
```

```python
# onehot_encoder.transform([['red1','large2','plastic'],['blue','small2','wood']]) 
# ValueError: Found unknown categories ['red1'] in column 0 during transform
```

```python
onehot_encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore') # Default handle_unknow is error
values=onehot_encoder.fit_transform(x_train)
onehot_encoder.transform([['red1','large3','plastic'],['blue','small','wood']])
```

```python
# red1 and large3 is not in the data so it gives 0 
```

```python
onehot_encoder.get_feature_names_out()
```

# Ordinal Encoder()

```python
ordinal_encoder=OrdinalEncoder()
```

```python
updated_xtrain=ordinal_encoder.fit_transform(x_train)
updated_xtrain
```

```python
# updated_xtest=ordinal_encoder.transform(xtest[['color']])
# It excepts 2D and also all the feature column names which are passed
```

```python
updated_xtest=ordinal_encoder.transform(x_test)
updated_xtest
```

```python
ordinal_encoder.get_feature_names_out()
```

```python
# To handle unknown values we need to give "use_encoded_values" and unknown_values =1
```

```python
ordinal_encoder=OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1)

updated_xtrain=ordinal_encoder.fit_transform(x_train)
updated_xtest=ordinal_encoder.transform(x_test)

values=ordinal_encoder.transform([['red1','large3','plastic'],['blue','small','wood']])
values
```

```python
# red1 and large3 are not in the data so ordinal encoder has assigned -1 to it
```

# Target Encoder

```python
target_encoder=TargetEncoder(cols=['color'],handle_unknown='value')
updated_xtrain=target_encoder.fit_transform(x_train,y_train)
updated_xtrain
```

```python
updated_xtest=target_encoder.transform(x_test)
updated_xtest
```
