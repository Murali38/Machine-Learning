# Indexing and Slicing

```python
# Indexing 
# We access the values by using +ve and -ve index
# If the index is not present it generates the error
# We can access the value by using step value also
# We can modify the values
```

```python
# Syntax -> arrayname[rowindex,columnindex]
# [start:stop:step(row),start:stop:step(col)]

arr[0]
```

```python
arr[-1]
```

```python
# arr[-4]
# IndexError: index -4 is out of bounds for axis 0 with size 3
```

```python
arr[2,2]
```

```python
arr[0,1]
```

```python
# Modifying

arr[0]=[11,22,33]
arr
```

```python
# Slicing

arr[1:3,1:3] # [row slic :column slice]

# ([[11, 22, 33],
# [ 4,  5,  6],
# [ 7,  8,  9]])

# from row 1 to row 3 (3 will be excluded) row 1,row 2
# from column 1 to column 3 (3 will be excluded) column 1,column 2
```

```python
arr[0:2,1:3]
# Here column starts with 1 
```

```python
arr[3:0,3:2]
```

# Variance and Covariance

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
n=len(x)
n
```

```python
var_x=np.sum((x-x.mean())**2)/n
# first it takes the mean of x and then it substracts the original x from mean x
# then it squares the x which is substracted from mean
# then it adds the values (sum)
# the it divide the values by n (len(n) = 5)
var_x
```

```python
# Instead of doing all the steps just use var method
np.var(x)
```

```python
var_y=np.sum((y-y.mean())**2)/n
var_y
```

```python
np.var(y)
```
